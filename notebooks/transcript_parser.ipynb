```python
import re

def is_timestamp(line):
    # Regular expression to match both "HH:MM:SS" and "MM:SS" formats
    # This pattern allows for "HH:MM:SS", "MM:SS", and also accounts for "H:MM:SS" and "M:SS"
    timestamp_pattern = r'^(\d{1,2}):[0-5]\d(:[0-5]\d)?$'
    return re.match(timestamp_pattern, line.strip())

def process_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            # Remove all instances of "[Music]"
            clean_line = line.replace("[Music]", "").strip()
            if not is_timestamp(clean_line) and clean_line:  # Check if the cleaned line is not a timestamp and not empty
                print(clean_line)

# Replace 'your_file.txt' with the path to your text file
process_file('keynote_transcript.txt')
```
